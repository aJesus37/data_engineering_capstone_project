```python
import pyspark.sql.functions as f
import sys
```

## Host Data

```python
df = spark.read.parquet('datalake/host_logs')
```

### Number of rows

```python
df.count()
```

### Log Schema

```python
df.printSchema()
```

```python
f = open('/dev/null', 'w')
old_stdout = sys.stdout
sys.stdout = f

df.unpersist()

sys.stdout = old_stdout
```
